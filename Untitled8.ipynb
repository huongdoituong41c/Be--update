```python
def minkowski_distance(x, y, r):
    """
    Tính Minkowski distance giữa hai điểm x và y với tham số r.

    Parameters:
    - x, y: Hai điểm trong không gian nhiều chiều (mảng NumPy).
    - r: Tham số trong công thức Minkowski distance.

    Returns:
    - Minkowski distance giữa p và q với tham số r.
    """
    distance = np.power(np.sum(np.power(np.abs(np.subtract(x, y)), r)), 1/r)
    return distance
```

```python
# Lấy dữ liệu iris
from sklearn import datasets
data = datasets.load_iris()
'''
Tạo một DataFrame target từ data.target, tức là nhãn của dữ liệu iris. Cột của DataFrame này được đặt tên là 'Label'.
Toàn bộ DataFrame được chỉ định một chỉ số dựa trên độ dài của data.target.
'''
target = pd.DataFrame({'Label': data.target.tolist()}, index=range(len(data.target.tolist())))
```

```python
'''
Thay thế các giá trị trong cột 'Label' của DataFrame target bằng các tên loài tương ứng.
Nó sử dụng phương thức replace() để thay thế 0 bằng 'Iris-Setosa', 1 bằng 'Iris-Versicolour', và 2 bằng 'Iris-Virginica'.
Tham số inplace=True chỉ định rằng thay đổi được thực hiện trực tiếp trên DataFrame target.
'''
target.replace({0: 'Iris-Setosa', 1: 'Iris-Versicolour', 2: 'Iris-Virginica'}, inplace=True)
```

```python
'''
Tạo một DataFrame features từ các cột của data.data.
Mỗi cột được đặt tên tương ứng với thuộc tính của hoa iris (sepals và petals).
Cột 'sepal_length' tương ứng với cột đầu tiên của data.data, và tương tự cho các cột khác.
'''
features = pd.DataFrame({'sepal_length': data.data[:,0].tolist(), 'sepal_width': data.data[:,1].tolist(),
                         'petal_length': data.data[:,2].tolist(), 'petal_width': data.data[:,-1].tolist()}, index=range(len(data.data[:, 0].tolist())))
```

```python
'''
Kết hợp features và target theo trục cột (axis=1).
df là DataFrame cuối cùng mà bạn có thể sử dụng cho việc phân tích dữ liệu.
Nó bao gồm tất cả các thuộc tính của hoa iris cùng với nhãn tương ứng.
'''
df = pd.concat([features, target], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y= df['Label']
# Chia dữ liệu thành 2 tập dữ liệu train và test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Hyperparameter
K = 70
# Tính khoảng cách
dis = []
for X_sub_train in X_train.to_numpy():
  md = minkowski_distance(X_test.to_numpy(), X_sub_train,2)
  dis.append(md)
```

```python
# sắp xếp tăng dần tập k gần nhất
k_indices = np.argsort(dis)[:K]
# k_nearest_labels Những phần tử k gần với labal nhất
k_nearest_labels = []
for i in k_indices.tolist():
    # Lấy giá trị y train, .tolist(): Chuyển đối tượng NumPy array thành một danh sách Python.
    Y_train = y_train.values.tolist()[i]
    # Lưu y train vào mảng k khoản cách nhãn
    k_nearest_labels.append(Y_train)
k_nearest_labels
```

```python
most_common = Counter(k_nearest_labels).most_common()
df_prediction = pd.concat([X_test, y_test],axis=1)
df_prediction['prediction'] = [most_common[0][0]] * df_prediction.shape[0]
df_prediction
```

```python
from sklearn.metrics import accuracy_score
# Tính độ chính xác của mô hình
acc = accuracy_score(y_test.to_numpy(), [most_common[0][0]] * y_test.to_numpy().shape[0])
acc
```
